```python
from nltk.tokenize import word_tokenize
```

```python
sent = 'Hello friends! How are you? Welcome to Python Programming.'
sent
```

## English

###Step 1: Remove Punctuations

```python
clean = [x for x in word_tokenize(sent) if x.isalpha() or x.isdigit()]
clean
```

```python
import string
```

```python
string.punctuation
```

```python
clean1 = [x for x in word_tokenize(sent) if x not in string.punctuation]
print(' '.join(clean1))
```

###Step 2: Remove StopWords

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
swords = stopwords.words('english')
swords, len(swords)
```

```python
clean_stopwords = [x for x in clean1 if x.lower() not in swords and x.isalnum()]
print(' '.join(clean_stopwords))
clean_stopwords
```
