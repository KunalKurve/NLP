## Porter Stemmer, Lancaster Stemming, Snowball Stemmer

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.stem import SnowballStemmer
```

```python
ps = PorterStemmer()
ls = LancasterStemmer()
ss = SnowballStemmer('english')
```

```python
ps.stem(words1[0]), ls.stem(words1[0]), ss.stem(words1[0])
```

```python
print("PorterStemmer")
for word in words1:
    print(word, '---->', ps.stem(word))

print("\nLancasterStemmer")
for word in words1:
    print(word, '---->', ls.stem(word))

print("\nSnowballStemmer")
for word in words1:
    print(word, '---->', ss.stem(word))
```

## With Sentence

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
sent = 'Hello friends! How are you? Welcome to Python Programming.'
sent
```

```python
swords = stopwords.words('english')
```

```python
clean = [x for x in word_tokenize(sent) if x.lower() not in swords and x.isalpha()]
clean
```

```python
for word in clean:
  print(word, '---->', ps.stem(word))
```
