```python
alphabets='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
int_to_char = dict(enumerate(alphabets))
```

```python
int_to_char
```

```python
char_to_int = dict((v,i) for i,v in enumerate(alphabets))
char_to_int
```

## Data Preparation

```python
seq_length = 1
x=[]
y=[]
for i in range(len(alphabets)-seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
    print(seq_in,'->',seq_out)
    
```

```python
from keras.utils import to_categorical
```

```python
y_new = to_categorical(y)

y_new
```

### Preapare Input Variables

```python
import numpy as np
```

```python
x = np.reshape(x, (25,1,1))
x
```

```python
x.shape
```

```python
x_new = x / 25

x_new
```

```python
x_new.shape, y_new.shape
```

### Create the model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(1,1), activation='relu'))
model.add(LSTM(32))
model.add(Dense(26, activation='softmax'))
```

### Compilt the Model

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Train the Model

```python
model.fit(x_new, y_new, epochs=1000, batch_size=1)
```

```python
model.evaluate(x_new,y_new,batch_size=1)
```

### Predict the character in sequence

```python
new = 'K'

new = char_to_int[new]
new = np.reshape(new,(1,1,1))
new = new/25
pred = model.predict(new)
pred = np.argmax(pred)
int_to_char[pred]
```

```python
seq_length = 3
x=[]
y=[]
for i in range(len(alphabets)-seq_length):
    seq_in = alphabets[i:i + seq_length]
    seq_out = alphabets[i + seq_length]
    x.append([char_to_int[char] for char in seq_in])
    y.append(char_to_int[seq_out])
    print(seq_in, '->', seq_out)
    
```

```python
x = np.reshape(x,(len(x),seq_length,1))
```

```python
x.shape
```

```python
x = x / 25
```

```python
model = Sequential()

model.add(Dense(25, input_shape= (3,1), activation='relu'))
model.add(LSTM(128))
model.add(Dense(26, activation='softmax'))
```

```python
y_new = to_categorical(y)
y_new.shape
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y_new, epochs=1000, batch_size=1)
```

```python
model.evaluate(x,y_new,batch_size=1)
```

### Predict on new sequence

```python
new = "BIG"
new = np.array([[char_to_int[x]] for x in new])
new = np.reshape(new,(1,3,1))
new = new / 25
pred = model.predict(new)
pred = np.argmax(pred)
int_to_char[pred]
```
