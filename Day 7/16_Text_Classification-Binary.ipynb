# Data Preparation

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
sent = 'Hello friends! How are you ? I like python programming.'
```

```python
# Step-1. Tokenize the text
tokens1 = word_tokenize(sent)
```

```python
tokens1
```

```python
# Step-2 Remove the punctuations

tokens2 = [token for token in tokens1 if token.isalnum()]
```

```python
tokens2
```

```python
# Step-3 Remove stopwords

swords = stopwords.words('english')
```

```python
tokens3 = [token for token in tokens2 if token.lower() not in swords]
```

```python
tokens3
```

```python
# step-4 Remove the suffixes
ps = PorterStemmer()

tokens4 = [ps.stem(token) for token in tokens3]
```

```python
tokens4
```

```python
def clean_txt(sent):
    tokens = word_tokenize(sent)  # Step-1. Tokenize the text
    tokens = [token for token in tokens if token.isalnum()]  # Step-2 Remove the punctuations
    tokens = [token for token in tokens if token.lower() not in swords]  # Step-3 Remove stopwords
    tokens = [ps.stem(token) for token in tokens]  # step-4 Remove the suffixes
    return tokens 
```

```python
sent = 'It was the expected result on that occasion.'
```

```python
clean_txt(sent)
```

```python
X.loc[400]
```

```python
clean_txt(X.loc[400])
```

```python
X.apply(clean_txt)
```

# TF*IDF Vectorize

```python
# Import the class
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Create the object
tfidf = TfidfVectorizer(analyzer=clean_txt)
```

```python
# Apply
X_new = tfidf.fit_transform(X)
```

```python
X_new
```

```python
X_new.toarray()
```

## Cross-Validation

```python
from sklearn.model_selection import train_test_split
```

```python
# seperate the training and testing data
X_train, X_test, y_train, y_test = train_test_split(X_new, y, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
tfidf.get_feature_names_out()
```

## Build the machine learning classification model

```python
# import the class
from sklearn.svm import SVC
```

```python
# Create the object
svm= SVC(kernel='sigmoid',random_state=0)
```

```python
%%time
svm.fit(X_train,y_train)
```

## Performance Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay
y_pred = svm.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred);
```
