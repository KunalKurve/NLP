2. Read the following file: https://www.london.ac.uk/sites/default/files/study-guides/introduction-to-natural-language-processing.pdf

Print only page number 12 and 13 from it.
Save this contents in a file nlp.txt

```python
!pip install PyMuPDF
import fitz
import requests
```

```python
pdf_url = "https://www.london.ac.uk/sites/default/files/study-guides/introduction-to-natural-language-processing.pdf"
pdf_filename = "nlp_guide.pdf"

pdf_response = requests.get(pdf_url)

with open(pdf_filename, "wb") as pdf_file:
    pdf_file.write(pdf_response.content)

doc = fitz.open(pdf_filename)
nlp_text = ""

for page_num in range(11, 13):  # pages 12 and 13 are indexed as 11 and 12
    page = doc.load_page(page_num)
    nlp_text += page.get_text()

# Save contents to nlp.txt
with open("nlp.txt", "w") as file:
    file.write(nlp_text)

# Print the extracted text for pages 12 and 13
print("Contents of pages 12 and 13 from the PDF:")
print(nlp_text)
```

3. Read the file from the link givcen below:
https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine- Learning-and-Natural-Language-Processing.docx

Read and display first 5 paragraphs from it. Count total ‘The’ present in the text.

```python
!pip install python-docx
from docx import Document
import requests

docx_url = "https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine-Learning-and-Natural-Language-Processing.docx"
docx_filename = "ml_nlp.docx"
```

```python
# Download the DOCX file
docx_response = requests.get(docx_url)
with open(docx_filename, "wb") as docx_file:
    docx_file.write(docx_response.content)
```

```python
# Read and display first 5 paragraphs
doc = Document(docx_filename)
first_5_paragraphs = []
for i in range(min(5, len(doc.paragraphs))):
    paragraph = doc.paragraphs[i]
    first_5_paragraphs.append(paragraph.text)
```

```python
# Count total 'The' present in the text
doc_text = "\n".join([paragraph.text for paragraph in doc.paragraphs])
count_the = doc_text.count('The')
print(f"Total occurrences of 'The' in the text: {count_the}")
```

```python
# Displaying first 5 paragraphs
print("First 5 paragraphs from the DOCX file:")
for i, paragraph in enumerate(first_5_paragraphs, 1):
    print(f"Paragraph {i}: {paragraph}")
```
