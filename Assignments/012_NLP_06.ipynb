```python
url_sri_lanka = "https://en.wikipedia.org/wiki/Sri_Lanka"
response = requests.get(url_sri_lanka)
html_content = response.text
html_content
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
text_content = soup.get_text()
text_content
```

```python
paragraphs = soup.find_all('p')
text = '\n'.join([para.text for para in paragraphs])
text
```

```python
# import parser and tokenizers
!pip install sumy
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer
```

```python
# Summarize using TextRank Summarizer


plain_parser = PlaintextParser.from_string(text, Tokenizer('english'))

#create object of TextRankSummarizer
text_rank_summarizer = TextRankSummarizer()

#Extract top 3 sentences
summary = text_rank_summarizer(plain_parser.document, sentences_count=3) #top 3 ranking sentences

for sent in summary:
    print(sent, '\n')
```

```python
from sumy.summarizers.lex_rank import LexRankSummarizer

lex_rank_summarizer = LexRankSummarizer()

summary = lex_rank_summarizer(plain_parser.document, sentences_count=3) #top 3 ranking sentences

for sent in summary:
    print(sent, '\n')
```

#### 3. LSA Summarizer

```python
from sumy.summarizers.lsa import LsaSummarizer

lsa_summarizer = LsaSummarizer()

summary = lsa_summarizer(plain_parser.document, sentences_count=3)

for sent in summary:
    print(sent, '\n')
```

4. Use following sentences to find the bag of words using count  

- The postman delivered the package to the wrong address.
- I wrapped a beautiful present for my friend's birthday.
- The delivery truck arrived late due to heavy traffic.
- We need to check the shipping address before sending the order.
- Online shopping offers a wide variety of products with fast delivery.

```python
from sklearn.feature_extraction.text import CountVectorizer


sentences = [
    "The postman delivered the package to the wrong address.",
    "I wrapped a beautiful present for my friend's birthday.",
    "The delivery truck arrived late due to heavy traffic.",
    "We need to check the shipping address before sending the order.",
    "Online shopping offers a wide variety of products with fast delivery."
]

cvt=CountVectorizer() # create obj of CountVectorizer

new_data=cvt.fit_transform(sentences) # transforms and stores counts

values=new_data.toarray()  # forms bag of words array

features=cvt.get_feature_names_out() #get the feature names

print("Bag of Words Array:",values)

print("\nFeature Names:",features)
```
