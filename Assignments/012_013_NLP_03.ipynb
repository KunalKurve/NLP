1. Apply chunking and show the chunks of the type VBD and DT

2. Find all the named entities present in article

```python
nlp = spacy.load('en_core_web_sm')
sent = nlp(text)
names = [str(x) for x in sent.ents]
names
```

3. Access all the words and apply Porter Stemmer and Lancaster Stemmer

```python
swords = stopwords.words('english')
Tokens = nltk.word_tokenize(text)
token1 = [token for token in Tokens if token.lower() not in swords]
words = [token.lower() for token in token1 if token.isalpha()]
words
```

```python
ps = PorterStemmer()
ls = LancasterStemmer()
```

```python
print("PorterStemmer")
for word in words:
    print(word, '---->', ps.stem(word))
```

```python
print("\nLancasterStemmer")
for word in words:
    print(word, '---->', ls.stem(word))
```

4. Access all the verbs in the past tense and lemmatize them. Show only unique verbs.

```python
wnl = WordNetLemmatizer()
verbs = set()
```

```python
for word in words:
    lemma = wnl.lemmatize(word, pos='v')
    if word != lemma and word.endswith('ed'):
        verbs.add(lemma)

for verb in verbs:
    print(verb)
```

5. Read only the adjectives from the text and find their synonyms.

```python
adjectives = set()

for word in words:
    lemma = wnl.lemmatize(word, pos='a')
    if word != lemma:
        adjectives.add(lemma)

for adj in adjectives:
    print(adj)
```

```python
synonyms = set()

for adj in adjectives:
  for syn in wordnet.synsets(adj):
    for lemma in syn.lemmas():
      synonyms.add(lemma.name())
  print(f"Synonyms for {adjectives}: {synonyms}")
```

6. Read all the verbs and print their antonyms.

```python
verbs = set()
for word, pos in nltk.pos_tag(words):
  if pos.startswith('VB'):
    lemma = wnl.lemmatize(word, pos='v')
    if word != lemma:
      verbs.add(lemma)
verbs
```

```python
antonyms = set()

for verb in verbs:
  for syn in wordnet.synsets(verb):
    for lemma in syn.lemmas():
      if lemma.antonyms():
        antonyms.add(lemma.name())
  print(f"Antonyms for {word}: {antonyms}")
```

7. Print all the entities marked as Date and Time in the data.

```python
nlp = spacy.load("en_core_web_sm")
text_string = " ".join(words)
text = nlp(text_string)
text
```

```python
for entity in text.ents:
    if entity.label_ in ["DATE", "TIME"]:
        print(entity.text, "-->", entity.label_)

date_time = set()

for entity in text.ents:
  if entity.label_ in ["DATE", "TIME"]:
    if entity.label not in date_time:
      date_time.add(entity.text)
      # print(entity.text, "-->", entity.label_)

date_time
```

8. Print all the name of location present in the article.

```python
loc = set()

for entity in text.ents:
  if entity.label_ == "GPE":
    if entity.label not in loc:
      loc.add(entity.text)
loc
```
