Q1. Copy the data about the NASA from wikipedia and perform the following:
1. Apply POS tagging on this data
2. Remove punctuation symbols
3. Delete the stop words
4. Perform morphological analysis on all the nouns
5. Find top three words in the data

```python
nasa = """The National Aeronautics and Space Administration (NASA; /ˈnæsə/) is an independent agency of the U.S. federal government responsible for the civil space program, aeronautics research, and space research.
 Established in 1958, it succeeded the National Advisory Committee for Aeronautics (NACA) to give the U.S. space development effort a distinct civilian orientation, emphasizing peaceful applications in space science.
 It has since led most of America's space exploration programs, including Project Mercury, Project Gemini, the 1968–1972 Apollo Moon landing missions, the Skylab space station, and the Space Shuttle.
 Currently, NASA supports the International Space Station (ISS) along with the Commercial Crew Program, and oversees the development of the Orion spacecraft and the Space Launch System for the lunar Artemis program."""
nasa
```

```python
#1. Apply POS tagging on this data
tags = pos_tag(word_tokenize(nasa))
tags
```

```python
# 2. Remove punctuation symbols
# clean_nasa = [x for x in word_tokenize(nasa) if x.isalpha() or x.isdigit()]
import string
clean_nasa = [word for word in word_tokenize(nasa) if word not in string.punctuation]
clean_nasa
```

```python
# 3. Delete the stop words
from nltk.corpus import stopwords
swords = stopwords.words('english')
clean_stopwords = [x for x in clean_nasa if x.lower() not in swords and x.isalnum()]
print(' '.join(clean_stopwords))
clean_stopwords
```

```python
# 4. Perform morphological analysis on all the nouns
from polyglot.text import Word

nouns = [t[0] for t in tags if t[1].startswith('N')]
nouns

for w in nouns:
  w = Word(w, language='en')
  print(w, '\t\t', w.morphemes)
```

```python
# 5. Find top three words in the data
words_freq = [x.lower() for x in clean_stopwords]
wfreq = {}
for word in words_freq:
    count = words_freq.count(word)
    wfreq[word] = count
sorted_wfreq = sorted(wfreq.items(), key=lambda x: x[1], reverse=True)
sorted_wfreq[:3]
```
